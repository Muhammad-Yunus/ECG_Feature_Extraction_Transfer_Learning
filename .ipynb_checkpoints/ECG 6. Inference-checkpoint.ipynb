### Inference Model Denoising & Classification


```python
# check versi Tensorflow & Keras
import tensorflow as tf
import keras 

print(tf.__version__)
print(keras.__version__)
```

```python
import os
import numpy as np
import pandas as pd
import datetime

from keras import models
from keras import backend as K
from tensorflow.keras import models as tf_kr_models
```

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.backends.backend_agg import FigureCanvasAgg
```

# load model

```python
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true))) 
```

```python
denoising_Conv_AE_model = models.load_model('denoising_conv_AE.h5', custom_objects={'rmse': rmse})
```

```python
cnn_model_Densenet = tf_kr_models.load_model('CNN_Classification_model_densenet121.h5')
cnn_model_Resnet = tf_kr_models.load_model('CNN_Classification_model_resnet50.h5')
cnn_model_VGG = tf_kr_models.load_model('CNN_Classification_model_vgg16.h5')
```

```python
label_names = ['AF', 'N']

fs = 250
```

```python
dataset_folder ="dataset/"
filenames = []
for filename in os.listdir(dataset_folder):
    if filename.find("AFDB_pasien_") > -1:
        filenames.append(filename)
```

```python
filenames
```

# load inference data

```python
# scale data from 0 to 1
def scaler(X):
    res = []
    for x in X :
        global_min = x.min()
        x = np.reshape(x, (2, 300))
        for i in range(len(x)):
            idx = np.max(np.nonzero(x[i]))
            x[i][idx+1:] = global_min
        x = np.reshape(x, (600))
        res.append((x - x.min())/(x.max() - x.min()))
    return np.array(res)
```

```python
inference_data = {}
for name in filenames:
    infer = pd.read_csv(dataset_folder + name, header=None)
    inference_data[name] = infer
```

```python
inference_data[filenames[0]].head()
```

```python
mat_infer_data = {}
for name in filenames:
    infer_df = inference_data[name]
    X = infer_df.values
    X = scaler(X)
    mat_infer_data[name] = X
    
```

# Apply denoising model to inference data

```python
Conv_AE_den_res = {}
for name in filenames:
    Conv_AE_den_res[name] = denoising_Conv_AE_model.predict(mat_infer_data[name].reshape(len(mat_infer_data[name]), 600, 1))
    print("denoising data - %s" % name)
```

```python
def plot_decoded_signal(decoded_signal, original_signal, name):
    time = np.arange(len(decoded_signal[0])) / fs
    plt.figure(figsize=(15,4))

    plt.subplot(1,2,1)
    plt.plot(decoded_signal[0])
    plt.plot(decoded_signal[1])
    plt.title("Denoised Signal - " + name)
    plt.legend(['channel 1', 'channel 2'])
    plt.grid()

    plt.subplot(1,2,2)
    plt.plot(original_signal[0])
    plt.plot(original_signal[1])
    plt.title("Original Signal")
    plt.legend(['channel 1', 'channel 2'])
    plt.grid()

    plt.tight_layout()
    plt.show()
```

- Check result

```python
name = filenames[8]
idx = 65

y = Conv_AE_den_res[name][idx]
x = mat_infer_data[name][idx]

signal = np.reshape(x, (2, 300))
decoded_signal = np.reshape(y, (2, 300))

plot_decoded_signal(decoded_signal, signal, name="Conv-AE")
```

# Convert matrix to Image

```python
def sequence_to_img(data):
    fig = Figure(figsize=(8, 8), dpi=28)
    canvas = FigureCanvasAgg(fig)

    ax = fig.add_subplot(111)
    ax.plot(data[0])
    ax.plot(data[1])
    ax.set_ylim(0,1)
    ax.axis('off')
    ax.margins(0)
    fig.tight_layout(pad=0)

    canvas.draw()
    buf = canvas.buffer_rgba()
    img = np.asarray(buf)[:, :, :3]
    return img
```

```python
print("----- Signal to Image -------")

Conv_AE_images = {}
for name in filenames :
    images_temp = []
    n = int(len(Conv_AE_den_res[name])//2)
    for i in range(len(Conv_AE_den_res[name][:n])):
        signal = np.reshape(Conv_AE_den_res[name][i], (2, 300))
        img = sequence_to_img(signal)

        images_temp.append(img)

    tm = datetime.datetime.now().strftime("%H:%M:%S")
    print('[%s] - finish processing %d sample for %s' % (tm, len(Conv_AE_den_res[key][:n]), name))
    Conv_AE_images[name] = np.array(images_temp)
```

# Apply to CNN model

```python
cnn_Densenet_classif_result = {}
cnn_Resnet_classif_result = {}
cnn_VGG_classif_result = {}
for name in filenames:
    densenet_res = []
    resnet_res = []
    vgg_res = []
    for img in Conv_AE_images[name]:
        img = img.reshape(1, img.shape[0], img.shape[1], img.shape[2])
        densenet_res.append(cnn_model_Densenet.predict(img))
        resnet_res.append(cnn_model_Resnet.predict(img))
        vgg_res.append(cnn_model_VGG.predict(img))
    
    cnn_Densenet_classif_result[name] = densenet_res
    cnn_Resnet_classif_result[name] = resnet_res
    cnn_VGG_classif_result[name] = vgg_res
    print("predicting data - %s" % name)
```

- Check result

```python
def plot_cnn_result(data, label='Deep-AE', confidence=0.0):
    confidence = confidence*100
    # show results
    data = data.reshape(2,300)
    time = np.arange(len(data[0])) / fs
    
    plt.figure(figsize=(6,4))
    plt.title("Classification Result : %s - confidence (%.2f%%)" % (label, confidence))
    plt.xlabel("samples - n")
    plt.plot(time, data[0],"g", label="channel 1")
    plt.plot(time, data[1],"r", label="channel 2")
    plt.tight_layout()
    plt.legend()
    plt.grid()
    plt.show()
```

# Result CNN + Densenet

```python
pred_result = []
for name in filenames :
    cnn_res = cnn_Densenet_classif_result[name]
    print("\n\n\nPrediction Result - %s\n" % name)
    for i, res in enumerate(cnn_res):
        predicted_id = np.argmax(res[0])
        label = label_names[predicted_id]
        confidence = res[0][predicted_id]
        label = label if confidence > 0.50 else '?' # threshold for confidence level , set to 70%
        pred_result.append([name, i, label, confidence])
        
        print('%s \t index - %d \t %s \t %.2f %%' % (name, i, label, confidence*100))
```

```python
pred_result_df = pd.DataFrame(pred_result, columns=['filename', 'idx', 'label', 'confidence'])
```

```python
pred_result_df.head()
```

```python
for name in filenames:
    print(name)
    count = dict(pred_result_df[pred_result_df['filename'] == name]['label'].value_counts())
    for key in count :
        print("%s \t: %d" % (key, count[key]))
    print('\n')
```

```python
def plot_donat(porsion, title, labels, colors=['red', 'green', 'blue']):
    plt.figure(figsize=(10,5))

    circle = plt.Circle( (0,0), 0.7, color='white')
    plt.pie(porsion, 
            labels=labels, 
            colors=colors,
            autopct='%1.1f%%')

    plt.title(title)
    p = plt.gcf()
    p.gca().add_artist(circle)
    plt.show()
```

```python
for name in filenames:
    count = dict(pred_result_df[pred_result_df['filename'] == name]['label'].value_counts())
    plot_donat(count.values(), name, count.keys())
```

- save result as csv

```python
pred_result_df.to_csv(dataset_folder + 'inference_result_cnn_Densenet_pasien_AF_dan_Normal.csv')
```
